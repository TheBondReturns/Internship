```python
import numpy as np

# (i) N = 1200
m = 1
N = 1200
k = 1/N


#below formulae implement a "negative order" diffraction
# negative signs already considered, so all outputs are positive

beta = lambda wavelength: np.arcsin((m*wavelength*(10**-6)/k)+np.sin(alpha))*180/np.pi #returns angle of diffraction in deg

wavelength =  lambda beta: k*(np.sin(beta*np.pi/180) - np.sin(alpha))/m * 10**6 #returns wavelength in nm, given angle of diff

# alpha = 0, first order diffraction with diffracted ray on the opposite side of incident ray

alpha = 0 * np.pi/180
print("N = 1200 , alpha = 0 :")
λmin=540
λmax=660
βmin=beta(λmin)
βmax=beta(λmax)
print("β(λ) min =",  βmin,",  β(λ) max =",  βmax,",  span =",βmax-βmin)


print("\n")

# alpha = 25.5 gives no rays in 540 to 660 range under first order diffraction on opposite side for valid angles.

# for the case of diffracted ray on the same side of normal, 
alpha = 25.5 * np.pi/180

#new formulae for same side diffraction, accounting for all signs:
beta = lambda wavelength: np.arcsin((m*wavelength*(10**-6)/k)-np.sin(alpha))*180/np.pi #returns angle of diffraction in deg

wavelength =  lambda beta: k*(np.sin(beta*np.pi/180) + np.sin(alpha))/m * 10**6 #returns wavelength in nm, given angle of diff

print("N = 1200 , alpha = 25.5 with Diffraction occuring on same side as incident ray")
λmin=540
λmax=660
βmin=beta(λmin)
βmax=beta(λmax)
print("β(λ) min =",  βmin,",  β(λ) max =",  βmax,",  span =",βmax-βmin)
```
